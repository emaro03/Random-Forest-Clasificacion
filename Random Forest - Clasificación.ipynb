## SELECCIÓN DE VARIABLES PREDICTORAS Y A PREDECIR

```python
columnas = df.columns.values.tolist()
predictoras = columnas[:3]

X = df[predictoras]
Y = df.Species
```

## IMPLEMENTACIÓN DEL ALGORITMO RANDOM FOREST REGRESION

```python
bosque = RandomForestClassifier(n_jobs = 2, oob_score = True, n_estimators = 500)
bosque.fit(X,Y)
```
